```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def lstm_step(x, h_prev, c_prev, params):

    f = sigmoid(params["wf"] * x + params["uf"] * h_prev + params["bf"])

    i = sigmoid(params["wi"] * x + params["ui"] * h_prev + params["bi"])

    c_tilde = math.tanh(params["wc"] * x + params["uc"] * h_prev + params["bc"])

    c = f * c_prev + i * c_tilde

    o = sigmoid(params["wo"] * x + params["uo"] * h_prev + params["bo"])

    h = o * math.tanh(c)

    return f, i, c_tilde, c, o, h


sequence = [1, 2, 3]

params = {
    "wf": 0.5, "uf": 0.1, "bf": 0,
    "wi": 0.6, "ui": 0.2, "bi": 0,
    "wc": 0.7, "uc": 0.3, "bc": 0,
    "wo": 0.8, "uo": 0.4, "bo": 0
}

wy = 4
by = 0

h = 0
c = 0

print("LSTM Numerical Example")
print("-" * 40)

for t, x in enumerate(sequence, start=1):
    f, i, c_tilde, c, o, h = lstm_step(x, h, c, params)

    print(f"Time step {t}")
    print("x =", x)
    print("f =", round(f, 3))
    print("i =", round(i, 3))
    print("c_tilde =", round(c_tilde, 3))
    print("c =", round(c, 3))
    print("o =", round(o, 3))
    print("h =", round(h, 3))
    print()

x_next = 4
f, i, c_tilde, c, o, h = lstm_step(x_next, h, c, params)

y_hat = wy * h + by

print("Prediction step")
print("x =", x_next)
print("f =", round(f, 3))
print("i =", round(i, 3))
print("c_tilde =", round(c_tilde, 3))
print("c =", round(c, 3))
print("o =", round(o, 3))
print("h =", round(h, 3))
print("Predicted next value =", round(y_hat, 3))
```
